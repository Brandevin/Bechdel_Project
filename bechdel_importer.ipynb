```python
import pandas as pd
import numpy as np
import pathlib
```

```python
import urllib.request, json

url_site = 'http://bechdeltest.com/api/v1/getAllMovies'
with urllib.request.urlopen(url_site) as url:
    data = json.loads(url.read().decode())        
```

```python
bechdel=pd.DataFrame(data)
bechdel
```

```python
#bechdel['imdb']='tt'+(str(bechdel['id']).split('\n')[0]).replace(" ", "").zfill(7)
bechdel['imdb']=bechdel.apply(lambda row:'tt'+str(row['imdbid']).zfill(7),axis=1)
```

```python
def category(row):
    if row['rating']==0:
        return 'nowomen'
    if row['rating']==1:
        return 'notalk'
    if row['rating']==2:
        return 'men'
    if row['rating']==3:
        return 'ok'
    
def category2(row):
    if row['rating']==3:
        return 'PASS'
    else:
        return 'FAIL'
```

```python
bechdel['clean_test']=bechdel.apply(lambda row:category(row),axis=1)
```

```python
bechdel['binary']=bechdel.apply(lambda row:category2(row),axis=1)
```

```python
bechdel.drop(['id','imdbid','rating'],inplace=True, axis=1)
```
